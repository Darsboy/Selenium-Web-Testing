```python
from selenium import webdriver
import time
```

```python
# need to xattr -d com.apple.quarantine chromedriver
```

```python
d = webdriver.Firefox()
```

```python
d.get("https://www.zolo.ca/") 
d.set_window_size(1217, 5924)
d.save_screenshot('scs1.png')
```

```python
sineInbox  = d.find_element_by_xpath("/html/body/div[1]/header/nav/div[2]/div/nav/div[3]/button[2]")
sineInbox.click()
```

```python
emailBox = d.find_element_by_xpath('//*[@id="email"]')
emailBox.send_keys("darcy.ldx@gmail.com")
```

```python
submitEmail = d.find_element_by_xpath('//*[@id="submitEmail"]')
submitEmail.click()
```

```python
SearchBar = d.find_element_by_xpath('//*[@id="sarea"]')


SearchBar.clear()
SearchBar.send_keys('Newmarket')
d.find_element_by_xpath('/html/body/div[1]/header/nav/div[1]/div[2]/form/div/button/i').click()
```

```python
botton_ele = d.find_element_by_xpath('//*[@id="gallery"]')
total_width  = botton_ele.size["width"]+100
total_height = botton_ele.size["height"]
print(total_width)
print(total_height)
    #total_height = 8000
d.set_window_size(total_width, total_height)
d.save_screenshot('neisw-fullpage.png')
```
